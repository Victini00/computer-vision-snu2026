### 2.1. Two-view triangulation

````python
import pickle
import os
from pathlib import Path
import torch
import numpy as np

from utils import load_cam_data


def two_view_triangulation(camera_matrix_1, camera_matrix_2, points_1, points_2, score_1, score_2):
    """
        input
        camera_matrix_n: 3 X 4, camera matrix of each camera
        points_n: N x 2, 2D coordinate of N points

        output
        output_dict: dictionary of triangulated points
        key: point_idx from 0 to 18, value: triangulated point, np.array with shape 1 X 3
        output_dict = {
            0: [[x1, y1, z1]],
            1: [[x2, y2, z2]],
            ...
            18: [x18, y18, z18]
          }
    """

    CONFIDENCE_SCORE_THRESHOLD = 0.3

    output_dict = {}
    camera_matrix = np.stack([camera_matrix_1, camera_matrix_2])

    for i in range(points_1.shape[0]):

        if score_1[i] < CONFIDENCE_SCORE_THRESHOLD or score_2[i] < CONFIDENCE_SCORE_THRESHOLD:
            continue

        points = np.stack([points_1[i], points_2[i]])

        A = []

        for P, (u, v) in zip(camera_matrix, points):

            A.append(u * P[2] - P[0])
            A.append(v * P[2] - P[1])

        A = np.array(A)

        U, S, V_t = np.linalg.svd(A)

        X = V_t[-1] # V의 마지막 열

        temp_output_dict = X[:3] / X[3]
        output_dict[i] = temp_output_dict.reshape(1, 3)

    return output_dict

#################### DO NOT MODIFFY CODES BELOW ####################
data_dir = Path("/Users/kwonsinyoung/Documents/2026-1/CV/content/CV_HW4_tools/test_set") / data_name

# 0.0. load cam data
cam_pth = data_dir / 'floored_scaled_transform.npz'
assert cam_pth.exists(), f"Camera data not found at {str(cam_pth)}"
intr_pth = data_dir / 'undistorted_intrinsics.json'
assert intr_pth.exists(), f"Intrinsic data not found at {str(intr_pth)}"
cam_data = load_cam_data(cam_pth, intr_pth)

# 0.1. load keypoint data
keypoint_dir = Path("/Users/kwonsinyoung/Documents/2026-1/CV/content/CV_HW4_tools/test_set") / data_name / "dwpose" / "kpts"
kp_path_list = sorted(list(keypoint_dir.glob("*.pkl")))
kp_dict = {}

for next_kp_path in kp_path_list:
    cam_idx = int(os.path.basename(next_kp_path)[:-4])
    with open(next_kp_path, "rb") as f:
        k = pickle.load(f)
    body_kp = np.array(k['body_keypoints'][0])
    kp_dict[cam_idx] = body_kp

# 2. two_view_triangulation (Here we selected cam 7 and 27)
cam_mat_7 = cam_data[7]['intrinsics'] @ (np.linalg.inv(cam_data[7]['c2w'])[:3])
cam_mat_27 = cam_data[27]['intrinsics'] @ (np.linalg.inv(cam_data[27]['c2w'])[:3])

kp_7 = kp_dict[7][:, :2]
score_7 = kp_dict[7][:, 2]
kp_27 = kp_dict[27][:, :2]
score_27 = kp_dict[27][:, 2]

two_view_triangulated = two_view_triangulation(cam_mat_7, cam_mat_27, kp_7, kp_27, score_7, score_27)

np.save(data_dir / "two_view_triangulated.npy", two_view_triangulated)
#################### DO NOT MODIFFY CODES ABOVE ####################
````

### 2.2. N-view triangulation

````python
import pickle
import os
import torch
import numpy as np
from pathlib import Path

from utils import load_cam_data


def N_view_naive_triangulation(camera_matrices_N, points_N, scores_N):
    """
        input
        camera_matrices_N: N X 3 X 4, N camera matrices of each camera
        points_N: N X M X 2, M joint coordinate in 2D in N views
        scores_N: N X M, M joint score in N views

        output
        output_dict: dictionary of triangulated points
        key: point_idx from 0 to 18, value: triangulated point, np.array with shape 1 X 3
        output_dict = {
            0: [[x1, y1, z1]],
            1: [[x2, y2, z2]],
            ...
            18: [x18, y18, z18]
          }
    """

    CONFIDENCE_SCORE_THRESHOLD = 0.3

    N, M, _ = points_N.shape

    output_dict = {}

    for i in range(M):

        valid = np.where(scores_N[:, i] >= CONFIDENCE_SCORE_THRESHOLD)[0]

        if len(valid) < 2:
            continue

        A = []

        for index in valid:
            P = camera_matrices_N[index]
            u, v = points_N[index, i]

            A.append(u * P[2] - P[0])
            A.append(v * P[2] - P[1])

        A = np.array(A)

        U, S, V_t = np.linalg.svd(A)

        X = V_t[-1]

        temp_output_dict = X[:3] / X[3]
        output_dict[i] = temp_output_dict.reshape(1, 3)

    return output_dict

#################### DO NOT MODIFFY CODES BELOW ####################
data_dir = Path("/Users/kwonsinyoung/Documents/2026-1/CV/content/CV_HW4_tools/test_set") / data_name

# 0.0. load cam data
cam_pth = data_dir / 'floored_scaled_transform.npz'
assert cam_pth.exists(), f"Camera data not found at {str(cam_pth)}"
intr_pth = data_dir / 'undistorted_intrinsics.json'
assert intr_pth.exists(), f"Intrinsic data not found at {str(intr_pth)}"
cam_data = load_cam_data(cam_pth, intr_pth)

# 1. load keypoint data
keypoint_dir = Path("/Users/kwonsinyoung/Documents/2026-1/CV/content/CV_HW4_tools/test_set") / data_name / "dwpose" / "kpts"
kp_path_list = sorted(list(keypoint_dir.glob("*.pkl")))
kp_dict = {}

for next_kp_path in kp_path_list:
    cam_idx = int(os.path.basename(next_kp_path)[:-4])
    with open(next_kp_path, "rb") as f:
        k = pickle.load(f)
    body_kp = np.array(k['body_keypoints'][0])
    kp_dict[cam_idx] = body_kp

# 3. N_view_triangulation

cam_mat_total = []
for cam_idx in kp_dict.keys():
    value = cam_data[cam_idx]
    cam_mat_process = value['intrinsics'] @ (np.linalg.inv(value['c2w'])[:3])
    cam_mat_total.append(cam_mat_process)
cam_mat_total = np.array(cam_mat_total)

kp_total = []
score_total = []

for next_kp in kp_dict.values():
    kp_total.append(next_kp[:, :2])
    score_total.append(next_kp[:, 2])

kp_total = np.array(kp_total)
score_total = np.array(score_total)

N_view_triangulated = N_view_naive_triangulation(cam_mat_total, kp_total, score_total)

np.save(data_dir / "N_view_triagulated.npy", N_view_triangulated)
#################### DO NOT MODIFFY CODES ABOVE ####################
````

### 3.1. Two-view brute-force triangulation

````python
import pickle
import os
import torch
import numpy as np
from pathlib import Path

from utils import load_cam_data


def bruteforce_two_view(camera_matrices_2, points_2, scores_2):
    """
        input
        camera_matrices_2: 2 X 3 X 4, camera matrices of each camera
        points_2: 2 X 2 X M X 2, 2 people's M joint coordinate in 2D in 2 views
        scores_2: 2 X 2 X M, 2 people's M joint score in N views

        output
        output_dict: key consists of [0, 1], value consists of dictionary of each person's triangulated point like in the format of 2-1 and 2-2
        output_dict = {
          0: {
            0: [[x1, y1, z1]],
            1: [[x2, y2, z2]],
            ...
            18: [x18, y18, z18]
          }
          1: {
            ...
          }
        }

    """

    output_dict = {}

    CONFIDENCE_SCORE_THRESHOLD = 0.3

    P_0, P_1 = camera_matrices_2[0], camera_matrices_2[1]

    M = points_2.shape[2]

    best_error = float('inf')
    best_order = [0, 1] # init

    order_list = [[0, 1], [1, 0]]

    for order in order_list:

        person_errors = []

        for person_index in range(2):

            detection_index = order[person_index]
            joint_errors = []

            for i in range(M):

                if scores_2[person_index, 0, i] < CONFIDENCE_SCORE_THRESHOLD or \
                   scores_2[detection_index, 1, i] < CONFIDENCE_SCORE_THRESHOLD:
                    continue

                point_0, point_1 = points_2[person_index, 0, i], points_2[detection_index, 1, i]

                A = np.array([point_0[0] * P_0[2] - P_0[0],
                              point_0[1] * P_0[2] - P_0[1],
                              point_1[0] * P_1[2] - P_1[0],
                              point_1[1] * P_1[2] - P_1[1]]
                            )
                
                U, S, V_t = np.linalg.svd(A)

                X_h = V_t[-1]
                X_3d = X_h[:3] / X_h[3]

                repro_error_0 = P_0 @ np.append(X_3d, 1)
                repro_error_1 = P_1 @ np.append(X_3d, 1)

                repro_error_0 = repro_error_0[:2] / repro_error_0[2]
                repro_error_1 = repro_error_1[:2] / repro_error_1[2]

                # repro error의 평균: joint error
                joint_errors.append((np.linalg.norm(repro_error_0 - point_0) + np.linalg.norm(repro_error_1 - point_1)) / 2)
            
            if joint_errors:
                # joint의 평균: person error
                person_errors.append(np.mean(joint_errors)) 

        final_error = np.mean(person_errors) if person_errors else float('inf')

        if final_error < best_error:

            best_error = final_error
            best_order = order

    # 최종 triangulation
    for person_index in range(2):

        best_detection_index = best_order[person_index]

        output_dict[person_index] = {}

        for i in range(M):
            if scores_2[person_index, 0, i] < CONFIDENCE_SCORE_THRESHOLD or \
               scores_2[best_detection_index, 1, i] < CONFIDENCE_SCORE_THRESHOLD:
                continue

            point_0, point_1 = points_2[person_index, 0, i], points_2[best_detection_index, 1, i]

            A = np.array([point_0[0] * P_0[2] - P_0[0],
                          point_0[1] * P_0[2] - P_0[1],
                          point_1[0] * P_1[2] - P_1[0],
                          point_1[1] * P_1[2] - P_1[1]]
                        )
            
            U, S, V_t = np.linalg.svd(A)

            X_h = V_t[-1]
            X_3d = X_h[:3] / X_h[3]

            output_dict[person_index][i] = X_3d.reshape(1, 3)

    return output_dict

#################### DO NOT MODIFFY CODES BELOW ####################
data_dir = Path("/Users/kwonsinyoung/Documents/2026-1/CV/content/CV_HW4_tools/test_set") / data_name

# 0.0. load cam data
cam_pth = data_dir / 'floored_scaled_transform.npz'
assert cam_pth.exists(), f"Camera data not found at {str(cam_pth)}"
intr_pth = data_dir / 'undistorted_intrinsics.json'
assert intr_pth.exists(), f"Intrinsic data not found at {str(intr_pth)}"
cam_data = load_cam_data(cam_pth, intr_pth)

# 1. load hpe output
keypoint_dir = Path("/Users/kwonsinyoung/Documents/2026-1/CV/content/CV_HW4_tools/test_set") / data_name / "dwpose" / "kpts"

kp_path_list = sorted(keypoint_dir.glob("*.pkl"))

kp_dict = {}

for next_kp_path in kp_path_list:
    cam_idx = int(os.path.basename(next_kp_path)[:-4])
    with open(next_kp_path, "rb") as f:
        k = pickle.load(f)
    if len(k['body_keypoints']) < 2:
        continue
    body_kp_1 = np.array(k['body_keypoints'][0])
    body_kp_2 = np.array(k['body_keypoints'][1])
    kp_dict[cam_idx] = [body_kp_1, body_kp_2]

# 2. two_view_triangulation

cam_mat_9 = cam_data[9]['intrinsics'] @ (np.linalg.inv(cam_data[9]['c2w'])[:3])
cam_mat_24 = cam_data[24]['intrinsics'] @ (np.linalg.inv(cam_data[24]['c2w'])[:3])

cam_mat_two_view = np.stack([cam_mat_9, cam_mat_24])

kp_9_1 = kp_dict[9][0][:, :2]
kp_9_2 = kp_dict[9][1][:, :2]
score_9_1 = kp_dict[9][0][:, 2]
score_9_2 = kp_dict[9][1][:, 2]
kp_24_1 = kp_dict[24][0][:, :2]
kp_24_2 = kp_dict[24][1][:, :2]
score_24_1 = kp_dict[24][0][:, 2]
score_24_2 = kp_dict[24][1][:, 2]

kp_two_p_two_view = np.zeros((2, 2, 19, 2))
kp_two_p_two_view[0, 0] = kp_9_1
kp_two_p_two_view[0, 1] = kp_24_1
kp_two_p_two_view[1, 0] = kp_9_2
kp_two_p_two_view[1, 1] = kp_24_2

score_two_view = np.zeros((2, 2, 19))
score_two_view[0, 0] = score_9_1
score_two_view[0, 1] = score_24_1
score_two_view[1, 0] = score_9_2
score_two_view[1, 1] = score_24_2

two_view_brute = bruteforce_two_view(cam_mat_two_view, kp_two_p_two_view, score_two_view)

np.save(data_dir / "two_view_brute.npy", two_view_brute)
#################### DO NOT MODIFFY CODES ABOVE ####################
````

## 3.2. N-view brute triangulation

````python
import pickle
import os
import torch
import numpy as np
from pathlib import Path
from tqdm import tqdm

from utils import load_cam_data

def N_view_greedy(camera_matrices_N, points_N, scores_N):
    """
        input
        camera_matrices_N: N X 3 X 4, camera matrices of each camera
        points_N: 2 X N X M X 2, 2 people's M joint coordinate in 2D in N views
        scores_N: 2 X N X M, 2 people's M joint score in N views

        output
        output_dict: key consists of [0, 1], value consists of dictionary of each person's triangulated point like in the format of 2-1 and 2-2
        output_dict = {
          0: {
            0: [[x1, y1, z1]],
            1: [[x2, y2, z2]],
            ...
            18: [x18, y18, z18]
          }
          1: {
            ...
          }
        }
    """

    N_views, M = camera_matrices_N.shape[0], points_N.shape[2]

    output_dict = {}

    CONFIDENCE_SCORE_THRESHOLD = 0.3

    best_assign_index = [0] * N_views
    best_error = float('inf') 

    for i in range(10000):

        assign = [0] + list(np.random.randint(0, 2, size=N_views-1))

        person_errors = []

        for p in range(2):

            joint_errors = []

            for j in range(M):

                valid = []

                for v_index in range(N_views):
                    
                    real_detection_index = assign[v_index] if p == 0 else 1 - assign[v_index]

                    if scores_N[real_detection_index, v_index, j] >= CONFIDENCE_SCORE_THRESHOLD:
                        valid.append(v_index)

                if len(valid) < 2:
                    continue

                A = []
                points_valid = []

                for v_index in valid:

                    real_detection_index = assign[v_index] if p == 0 else 1 - assign[v_index]

                    P = camera_matrices_N[v_index]
                    point = points_N[real_detection_index, v_index, j]

                    u, v = point

                    A.append(u * P[2] - P[0])
                    A.append(v * P[2] - P[1])

                    points_valid.append((P, point))

                U, S, V_t = np.linalg.svd(np.array(A))

                X_h = V_t[-1]
                X_3d = X_h[:3] / X_h[3]

                errors = []

                for P, point in points_valid:

                    repro_error = P @ np.append(X_3d, 1)
                    repro_error = repro_error[:2] / repro_error[2]

                    errors.append(np.linalg.norm(repro_error - point))

                joint_errors.append(np.mean(errors))

            if joint_errors:
                person_errors.append(np.mean(joint_errors))

        final_error = np.mean(person_errors) if person_errors else float('inf')

        if final_error < best_error:
            
            best_error = final_error
            best_assign_index = assign

    # 최종 triangulation
    for p in range(2):

        output_dict[p] = {}

        for j in range(M):

            valid = []

            for v_index in range(N_views):
                
                real_detection_index = best_assign_index[v_index] if p == 0 else 1 - best_assign_index[v_index]

                if scores_N[real_detection_index, v_index, j] >= CONFIDENCE_SCORE_THRESHOLD:
                    valid.append((v_index, real_detection_index))

            if len(valid) < 2:
                continue

            A = []

            for v_index, real_detection_index in valid:

                P = camera_matrices_N[v_index]

                u, v = points_N[real_detection_index, v_index, j]

                A.append(u * P[2] - P[0])
                A.append(v * P[2] - P[1])

            U, S, V_t = np.linalg.svd(np.array(A))

            X_h = V_t[-1]
            X_3d = X_h[:3] / X_h[3]

            output_dict[p][j] = X_3d.reshape(1, 3)

    return output_dict


#################### DO NOT MODIFFY CODES BELOW ####################
data_dir = Path("/Users/kwonsinyoung/Documents/2026-1/CV/content/CV_HW4_tools/test_set") / data_name

# 0.0. load cam data
cam_pth = data_dir / 'floored_scaled_transform.npz'
assert cam_pth.exists(), f"Camera data not found at {str(cam_pth)}"
intr_pth = data_dir / 'undistorted_intrinsics.json'
assert intr_pth.exists(), f"Intrinsic data not found at {str(intr_pth)}"
cam_data = load_cam_data(cam_pth, intr_pth)

# 1. load hpe output
keypoint_dir = Path("/Users/kwonsinyoung/Documents/2026-1/CV/content/CV_HW4_tools/test_set") / data_name / "dwpose" / "kpts"

kp_path_list = sorted(keypoint_dir.glob("*.pkl"))

kp_dict = {}

for next_kp_path in kp_path_list:
    cam_idx = int(os.path.basename(next_kp_path)[:-4])
    with open(next_kp_path, "rb") as f:
        k = pickle.load(f)
    if len(k['body_keypoints']) < 2:
        continue
    body_kp_1 = np.array(k['body_keypoints'][0])
    body_kp_2 = np.array(k['body_keypoints'][1])
    kp_dict[cam_idx] = [body_kp_1, body_kp_2]

cam_mat_total = []
for cam_idx in kp_dict.keys():
    value = cam_data[cam_idx]
    cam_mat_process = value['intrinsics'] @ (np.linalg.inv(value['c2w'])[:3])
    cam_mat_total.append(cam_mat_process)
cam_mat_total = np.array(cam_mat_total)

kp_total = np.zeros((2, len(cam_mat_total), 19, 2))
score_total = np.zeros((2, len(cam_mat_total), 19))

for view_idx, next_kp in enumerate(kp_dict.values()):
    kp_1 = next_kp[0][:, :2]
    kp_2 = next_kp[1][:, :2]
    score_1 = next_kp[0][:, 2]
    score_2 = next_kp[1][:, 2]
    kp_total[0, view_idx] = kp_1
    kp_total[1, view_idx] = kp_2
    score_total[0, view_idx] = score_1
    score_total[1, view_idx] = score_2

N_view_greedy_output = N_view_greedy(cam_mat_total, kp_total, score_total)

np.save(data_dir / "N_view_brute.npy", N_view_greedy_output)
#################### DO NOT MODIFFY CODES ABOVE ####################
````

### 4.1. Node(Joint) Proposals


Note that the heatmap resolution is 1/4 times of original image resolution on both X and Y axis. We provide you NMS function, but it's totally free to implement your own function. Note that the output is numpy array format here.

````python
from pathlib import Path


import torch
import torch.nn.functional as F
import numpy as np

from utils import load_cam_data, N_JNTS


def compute_3d_node_proposals(n_jnts, heatmaps, extrinsics, intrinsics, voxel_grid, transform_matrix, threshold=4.0):
    """
    input
      heatmaps: Dict[cam_id, Tensor(N,J,h,w)]
            - N: # of detected person
            - J: # of joints (19, ==N_JNTS)
            - h: heatmap height (about 375) (1/4 size of original image height)
            - w: heatmap width (about 500) (1/4 size of original image width)
      extrinsics: Dict[cam_id, 4x4 tensor (cam to world)]
      intrinsics: Dict[cam_id, 3x3 tensor]
            - Note that this intrinsic is for image sized about (1500, 2000) (slight different per view). you should handle resolution difference properly
      voxel_grid: Tensor(Gx, Gy, Gz, 3)
            - each voxel holds 3D world coordinate.
      threshold: float, score threshold for NMS

    output
      node_proposals: Dict[joint_idx, np.array(N_j, 3)] Dictionary of Numpy array
        - joint_idx(int): (0~18)
        - N_j: # of proposed nodes corresponding to j-th joint.

    """

    gx, gy, gz, _ = voxel_grid.shape

    # homogeneous coordinate 변환
    voxel_grid_flat = voxel_grid.reshape(-1, 3).float().cpu()
    voxel_homo = torch.cat([voxel_grid_flat, torch.ones(voxel_grid_flat.shape[0], 1)], dim=-1)

    node_proposals = dict() # dict 초기화

    for i in range(n_jnts):

        score_map = torch.zeros(gx * gy * gz)

        for cam_id, heatmap in heatmaps.items():

            N_det, J, h, w = heatmap.shape

            heatmap_j = heatmap[:, i:i+1, :, :].float()

            # 변환 행렬
            w2c = torch.linalg.inv(extrinsics[cam_id].float().cpu())
            K = intrinsics[cam_id].float().cpu()

            # 투영
            points_camera = (w2c @ voxel_homo.T).T[:, :3]

            valid = points_camera[:, 2] > 0

            projection = (K @ points_camera.T).T
            # 히트맵 크기는 원본의 1/4
            full_h = 4*h
            full_w = 4*w

            proj_xy = projection[:, :2] / projection[:, 2:3].clamp(min=1e-8)
            x_norm = (proj_xy[:, 0] / (full_w - 1)) * 2 - 1
            y_norm = (proj_xy[:, 1] / (full_h - 1)) * 2 - 1

            grid = torch.stack([x_norm, y_norm], dim = -1).unsqueeze(0).unsqueeze(0).expand(N_det, -1, -1, -1)

            sample = F.grid_sample(heatmap_j, grid,
                                   mode = 'bilinear', padding_mode='zeros',
                                   align_corners=True
                                   )

            camera_score = sample.squeeze(1).squeeze(1).max(dim=0).values * valid.float()

            score_map += camera_score

        # NMS 적용
        score_3d = score_map.reshape(gx, gy, gz)

        voxel_index, _ = nms3d(score_3d, tau = threshold)

        if voxel_index.numel() == 0:
            continue

        world_coords = voxel_grid.cpu()[voxel_index[:, 0], voxel_index[:, 1], voxel_index[:, 2]]

        node_proposals[i] = world_coords.numpy()

    return node_proposals


def nms3d(score_map, tau=0.3, radius=5):
    """
    Non-Maxima Suppression on a 3-D voxel score-map.

    Args:
        score_map (torch.Tensor): (Gx, Gy, Gz) tensor with 3-D node scores.
        tau (float)              : keep only peaks whose score > tau.
        radius (int)             : NMS window radius ⇒ window = (2r+1)^3.

    Returns:
        voxel_idx (N×3 long) — indices in the voxel grid (z,y,x).
        peak_scores (N,)     — their scores.
    """

    # (1,1,D,H,W) so we can use max_pool3d
    s = score_map.unsqueeze(0).unsqueeze(0)

    # local max within a cubic window
    pooled = F.max_pool3d(
        s, kernel_size=2 * radius + 1, stride=1, padding=radius
    )

    keep = (s == pooled) & (s > tau)                 # boolean mask for peaks
    voxel_idx = keep.nonzero(as_tuple=False)[:, -3:]  # drop batch/channel dims

    if voxel_idx.numel() == 0:
        return (
            voxel_idx,
            torch.empty((0,), device=score_map.device),
        )

    peak_scores = score_map[voxel_idx[:, 0],
                            voxel_idx[:, 1],
                            voxel_idx[:, 2]]

    return voxel_idx, peak_scores



def make_voxel_grid(c2ws, grid_size=0.02, device='cuda'):
    """
    Create a 3-D voxel grid in world coordinates.

    1. From camera locations, define large 3D box including all cameras with enough margins.
    2. Then divide the 3D box into small voxel cubes whose grid_size is as given. (0.02m == 2cm)

    Note that z=0 is ground and -z is up direction in calibrated cameras here! We HIGHLY recommend
    you to implement the box include all  -0.1 < z < 2.0  (as person height < 2.0m in general).


    Args:
        c2ws (Dict[int, torch.Tensor]): {cam_id: 4×4 cam-to-world matrix}.
        grid_size (float): edge length of each voxel (meters).
        device (str): 'cuda' or 'cpu'.

    Returns:
        voxel_grid (torch.Tensor): (Nx,Ny,Nz,3) world-space coordinates.
        transform_matrix (torch.Tensor): 4×4 matrix that maps voxel indices → world coords.
    """

    DEVICE = 'cpu' # mac

    camera_positions = torch.stack([c2w[:3, 3] for c2w in c2ws.values()]).float()

    min_xy = camera_positions[:, :2].min(dim=0).values - 0.5
    max_xy = camera_positions[:, :2].max(dim=0).values + 0.5

    min_z = -2.0
    max_z = 0.1

    x_coords = torch.arange(float(min_xy[0]), float(max_xy[0]), grid_size, device=DEVICE)
    y_coords = torch.arange(float(min_xy[1]), float(max_xy[1]), grid_size, device=DEVICE)
    z_coords = torch.arange(float(min_z), float(max_z), grid_size, device=DEVICE)

    grid_x, grid_y, grid_z = torch.meshgrid(x_coords, y_coords, z_coords, indexing='ij')
    voxel_grid = torch.stack([grid_x, grid_y, grid_z], dim=-1)

    # voxel -> world 좌표 변환
    transform_matrix = torch.eye(4, dtype=torch.float64, device=DEVICE)

    transform_matrix[0, 0] = grid_size
    transform_matrix[1, 1] = grid_size
    transform_matrix[2, 2] = grid_size
    
    transform_matrix[0, 3] = float(min_xy[0])
    transform_matrix[1, 3] = float(min_xy[1])
    transform_matrix[2, 3] = float(min_z)

    return voxel_grid, transform_matrix




#################### DO NOT MODIFFY CODES BELOW ####################
data_dir = Path("/Users/kwonsinyoung/Documents/2026-1/CV/content/CV_HW4_tools/test_set") / data_name

# 0.0. load heatmaps
heatmap_dir = data_dir / "dwpose" / 'heatmaps'
assert heatmap_dir.exists(), f"Heatmap data not found at {str(heatmap_dir)}"
heatmap_fdict = dict()
for heatmap_fname in heatmap_dir.glob("*.npy"):
    cam_id = int(heatmap_fname.stem.split('.')[0])
    heatmap_fdict[cam_id] = heatmap_fname

# 0.1. get cameras
cam_pth = data_dir / 'floored_scaled_transform.npz'
assert cam_pth.exists(), f"Camera data not found at {str(cam_pth)}"
intr_pth = data_dir / 'undistorted_intrinsics.json'
assert intr_pth.exists(), f"Intrinsic data not found at {str(intr_pth)}"
cam_data = load_cam_data(cam_pth, intr_pth)
extrinsics = {cam_id: torch.from_numpy(cam_data[cam_id]['c2w']) for cam_id in sorted(list(cam_data.keys()))}
intrinsics = {cam_id: torch.from_numpy(cam_data[cam_id]['intrinsics']) for cam_id in sorted(list(cam_data.keys()))}

# 1. get voxel grid of scene
voxel_grid, transform_matrix = make_voxel_grid(c2ws=extrinsics)

# 2. compute 3D node proposals
body_heatmap_dict = {cam_id: torch.from_numpy(np.load(heatmap_fdict[cam_id])[:N_JNTS]) for cam_id in heatmap_fdict}   # Use 19 body joints

node_proposals = compute_3d_node_proposals(
    n_jnts=N_JNTS,  # Number of joints in the body
    heatmaps=body_heatmap_dict,
    extrinsics=extrinsics,
    intrinsics=intrinsics,
    voxel_grid=voxel_grid,
    transform_matrix=transform_matrix,
)
print(f"Node proposals computed for {len(node_proposals)} joints.")

save_dir = data_dir / "panoptic"
save_dir.mkdir(parents=True, exist_ok=True)
np.save(save_dir / "node_proposals.npy", node_proposals)

#################### DO NOT MODIFFY CODES ABOVE ####################
````

### 4.2. Part(Bone) Proposals

Note that the heatmap resolution is 1/4 times of original image resolution on both X and Y axis. Also be aware that output part score list has numpy array as element.

````python
import pickle
from pathlib import Path

import torch
import torch.nn.functional as F
import numpy as np


from utils import load_cam_data, get_bones, create_bones


def generate_part_proposals(node_proposals, joint_pairs, heatmaps, extrinsics, intrinsics, threshold=0.05):
    """
    input
      node_proposals: Dict[joint_idx, Tensor(N_j, 3)]
        - joint_idx(int): (0~18)
        - N_j: # of proposed nodes corresponding to j-th joint.
      joint_pairs: [(u,v),...] bones defined by joint indices(int) of each bone.
      heatmaps: Dict[cam_id, Tensor(N,J,h,w)]
            - N: # of detected person
            - J: # of joints (19, ==N_JNTS)
            - h: heatmap height (=375) (1/4 size of original image height)
            - w: heatmap width (=502) (1/4 size of original image width)
      extrinsics: Dict[cam_id, 4x4 tensor (cam to world)]
      intrinsics: Dict[cam_id, 3x3 tensor]
            - Note that this intrinsic is for image sized (1500, 2008). you should mathc them properly
    output
      part_scores: List[np.array(N_u, N_v)] List of score maps
        - N_u: # of node proposals of u-th joint
        - N_v: # of node proposals of v-th joint

    """

    part_scores = []

    for u, v in joint_pairs:

        if (u not in node_proposals) or (v not in node_proposals):
            part_scores.append(np.zeros((0, 0)))
            continue

        nodes_u = node_proposals[u]
        nodes_v = node_proposals[v]

        N_u, N_v = len(nodes_u), len(nodes_v)

        score_matrix = np.zeros((N_u, N_v))
        n_valid_cams = 0

        for cam_id, heatmap in heatmaps.items():

            N_det, J, h, w = heatmap.shape
            heatmap = heatmap.float()

            hm_shape = torch.tensor([h, w])
            w2c = torch.linalg.inv(extrinsics[cam_id].float())
            K = intrinsics[cam_id].float()

            # heatmap에 투영, score 샘플링
            hm_u = heatmap[:, u:u+1, :, :]
            hm_v = heatmap[:, v:v+1, :, :]

            u_2d = project_points(nodes_u.float(), K, w2c, hm_shape)
            v_2d = project_points(nodes_v.float(), K, w2c, hm_shape)

            grid_u = u_2d.unsqueeze(0).unsqueeze(0).expand(N_det, -1, -1, -1)
            grid_v = v_2d.unsqueeze(0).unsqueeze(0).expand(N_det, -1, -1, -1)

            scores_u = F.grid_sample(hm_u, grid_u, mode='bilinear', padding_mode='zeros', align_corners=True)
            scores_u = scores_u.squeeze(1).squeeze(1).max(dim=0).values.detach().numpy()

            scores_v = F.grid_sample(hm_v, grid_v, mode='bilinear', padding_mode='zeros', align_corners=True)
            scores_v = scores_v.squeeze(1).squeeze(1).max(dim=0).values.detach().numpy()

            score_matrix += (scores_u[:, None] + scores_v[None, :]) / 2
            n_valid_cams +=1

        if n_valid_cams >0:
            score_matrix /= n_valid_cams

        distance = torch.norm(nodes_u[:, None, :]- nodes_v[None, :, :], dim=2).numpy()
        score_matrix *= np.exp(-distance / 0.1)

        part_scores.append(score_matrix)

    return part_scores  # List of numpy arrays for each joint pair





def project_points(points, K, E, hm_shape):
    """
    input
      points: Tensor(N, 3) 3D world coordinates
      K: Tensor(3,3) Intrinsic
      E: Tensor(4,4) Extrinsic (world to cam)
      hm_shape: Tensor(2): heatmap size (h, w)

    output:
      norm_2d: Tensor(N, 2) normalized 2D coordiante of projected keypoint in [-1, 1].

    As we provide the heatmap as 1/4 resized heatmap, we need to this customized
    function to get accurate normalized 2D coordinate of projected keypoints. It's
    fine to use your own function.
    """
    points_homo = torch.cat([points, torch.ones(*points.shape[:-1], 1, device=points.device)], dim=-1)
    points_cam = (E @ points_homo.view(-1,4).T).T[:, :3]
    proj_2d = (K @ points_cam.T).T
    proj_2d = proj_2d[:, :2] / proj_2d[:, 2:3]

    H, W = hm_shape
    H = H * 4
    W = W * 4

    norm_2d = proj_2d.clone()
    norm_2d[:, 0] = (norm_2d[:, 0] / (W - 1)) * 2 - 1
    norm_2d[:, 1] = (norm_2d[:, 1] / (H - 1)) * 2 - 1

    return norm_2d





#################### DO NOT MODIFFY CODES BELOW ####################
data_dir = Path("/Users/kwonsinyoung/Documents/2026-1/CV/content/CV_HW4_tools/test_set") / data_name

# 0.0. load dwpose outputs
heatmap_dir = data_dir / "dwpose" / 'heatmaps'
assert heatmap_dir.exists(), f"Heatmap data not found at {str(heatmap_dir)}"
heatmap_fdict = dict()
for heatmap_fname in heatmap_dir.glob("*.npy"):
    cam_id = int(heatmap_fname.stem.split('.')[0])
    heatmap_fdict[cam_id] = heatmap_fname

# 0.1. get cameras
cam_pth = data_dir / 'floored_scaled_transform.npz'
assert cam_pth.exists(), f"Camera data not found at {str(cam_pth)}"
intr_pth = data_dir / 'undistorted_intrinsics.json'
assert intr_pth.exists(), f"Intrinsic data not found at {str(intr_pth)}"
cam_data = load_cam_data(cam_pth, intr_pth)

extrinsics = {cam_id: torch.from_numpy(cam_data[cam_id]['c2w']) for cam_id in sorted(list(cam_data.keys()))}
intrinsics = {cam_id: torch.from_numpy(cam_data[cam_id]['intrinsics']) for cam_id in sorted(list(cam_data.keys()))}


# 0.2. load node proposals
node_proposals = np.load(data_dir / "panoptic" / "node_proposals.npy", allow_pickle=True).item()
node_proposals = {int(k): torch.tensor(v, dtype=torch.float32) for k, v in node_proposals.items()}

# 0.3. get bone connections
bones = get_bones()

# 1. Get Part Proposals
part_scores = generate_part_proposals(
    node_proposals=node_proposals,
    joint_pairs=bones,
    heatmaps={cam_id: torch.from_numpy(np.load(heatmap_fdict[cam_id])) for cam_id in heatmap_fdict},
    extrinsics=extrinsics,
    intrinsics=intrinsics,
    threshold=0.05
)

# save part proposals
save_dir = data_dir / "panoptic"
with open(save_dir / 'part_scores.pkl', 'wb') as f:
    pickle.dump(part_scores, f)

# save bones for viz
viz_parts = create_bones(
    part_scores=part_scores,
    joint_pairs=bones,
    node_proposals=node_proposals,
    viz_thrs=0.01
)
np.save(save_dir / "viz_part_proposals.npy", viz_parts)
#################### DO NOT MODIFFY CODES ABOVE ####################
````

### 4.3. Skeletal(person bone cluster) Proposals

The given code is just guidance. feel free to modify all except ```build_skeletons_with_dp_and_projection_filter``` function input output.


````python
import pickle
from pathlib import Path

import torch
import numpy as np
from tqdm import trange
from collections import defaultdict

from utils import load_cam_data, load_body_kpts, get_bones, create_bone_mesh

def build_skeletons_with_dp_and_projection_filter(
    joint_pairs,
    part_scores,
    node_proposals,
    body_kpts_dict,
    intrinsics,
    extrinsics,
    root_node_idx=1,        # Assuming root node is 1 (neck)
    min_joints=5,           # Discard skeleton less than 5 joints
    max_skeletons=10,       # To avoid infinite-loop, set upper bound on # of skeleton
    pixel_thresh=10.0,      # 10px in original paper, but VGA wile here is 2k so use 25.0 (fine to adjust)
):
    """
    input
      joint_pairs: [(u,v),...] bones defined by joint indices(int) of each bone.
      part_scores: List[Tensor(N_u, N_v)] List of score maps
        - N_u: # of node proposals of u-th joint
        - N_v: # of node proposals of v-th joint
      node_proposals: Dict[joint_idx, Tensor(N_j, 3)]
        - joint_idx(int): (0~18)
        - N_j: # of proposed nodes corresponding to j-th joint.
      body_kpts_dict: Dict[cam_id, Tensor(N, J, 3)]
        - keypoint is [x, y, s] where s denotes score.
      extrinsics: Dict[cam_id, 4x4 tensor (cam to world)]
      intrinsics: Dict[cam_id, 3x3 tensor]
            - Note that this intrinsic is for image sized (1500, 2008). you should mathc them properly
    output
      skeletons: List[Dict[joint_idx, np.array(3)]] list of clustered skeleton
      kpts_2ds: correspondence of skeleton 3D joints and 2d keypoint detection.
                define by yourself


    """



    # Define Skeleton Hierarchy for DP
    children = defaultdict(list)
    for i, (u, v) in enumerate(joint_pairs):
        children[u].append((v, i))

    skeletons = []
    kpts_2ds = []


    def assign_to_views(joint_3d, joint_idx):
        '''
        
        '''

        count = 0
        used_views = {}

        for cam_id, kpts in body_kpts_dict.items():

            K = intrinsics[cam_id].float()
            w2c = torch.linalg.inv(extrinsics[cam_id].float())

            point = joint_3d.float().cpu()

            point_homo = torch.cat([point, torch.tensor([1.0])], dim=0)
            point_camera = (w2c.cpu() @ point_homo)[:3]

            if point_camera[2] <= 0:
                continue

            point_image = K.cpu() @ point_camera
            point_2d = point_image[:2] / point_image[2]

            kpts_2d_camera = kpts[:, joint_idx, :2].float()

            distances = torch.norm(kpts_2d_camera - point_2d.unsqueeze(0), dim=1)

            best_det = int(distances.argmin().item())

            if distances[best_det] < pixel_thresh:

                count += 1

                used_views[cam_id] = (best_det, kpts[best_det, joint_idx, :2].numpy())

        
        return count, used_views


    dp = {}

    def DP(u,i):
        
        if (u, i) in dp:
            return dp[(u, i)]

        joint_u = node_proposals[u][i].float().cpu()

        u_view_count, _ = assign_to_views(joint_u, u)

        if u_view_count < 2:
            dp[(u, i)] = (0.0, {})
            return (0.0, {}) # fix

        # init
        total_score = 0.0
        assignments = {u: (joint_u, i)}

        for v, part_index in children.get(u, []):

            if v not in node_proposals or len(node_proposals[v]) == 0:
                continue

            scores = part_scores[part_index]

            if i >= len(scores):
                continue

            # init
            best_score_child = -np.inf
            best_assign_child = {}

            for j in range(len(node_proposals[v])):

                part_score = float(scores[i][j]) if j < scores.shape[1] else 0.0

                score_child, assign_child = DP(v, j)

                mix = part_score + score_child

                if mix > best_score_child:

                    best_score_child = mix
                    best_assign_child = assign_child

            if best_score_child > 0:

                total_score += best_score_child
                assignments.update(best_assign_child)

        dp[(u, i)] = (total_score, assignments)
        return (total_score, assignments)


    # We may define a new node, include root node in DP, and simplify code more.
    # It's up to you. the code below is just guidance

    used_root_indices = set()
    
    while len(skeletons) < max_skeletons:
        _v, part_idx = children.get(root_node_idx)[0]
        n_root = part_scores[part_idx].shape[0]       # number of root joints

        best_score = -np.inf
        best_root_idx = None
        nodes = None
        dp = {}

        for i in range(n_root):
            if i in used_root_indices:
                continue
            scores, parts = DP(root_node_idx, i)
            if scores > best_score:
                best_score = scores
                best_root_idx = i
                nodes = parts

        if nodes is None:
            break

        if len(nodes) < min_joints:
            break

        kpts_2d = {}

        for k_joint_index, (k_joint_3d, _) in nodes.items():

            _, viewmatch = assign_to_views(k_joint_3d, k_joint_index)

            kpts_2d[k_joint_index] = viewmatch

        used_root_indices.add(best_root_idx)

        skeletons.append({idx: pos.cpu().numpy() for idx, (pos, _) in nodes.items()})
        kpts_2ds.append(kpts_2d)

    return skeletons, kpts_2ds



#################### DO NOT MODIFFY CODES BELOW ####################
data_dir = Path("/Users/kwonsinyoung/Documents/2026-1/CV/content/CV_HW4_tools/test_set") / data_name
max_skeleton = 10

# 0.0. load body pose detections (In original camera space)
kpt_dir = data_dir / "dwpose" / 'kpts'
body_kpts_dict = load_body_kpts(kpt_dir)
body_kpts_dict = {cam_id: torch.from_numpy(body_kpts_dict[cam_id]) for cam_id in body_kpts_dict}

# 0.1. get cameras
cam_pth = data_dir / 'floored_scaled_transform.npz'
assert cam_pth.exists(), f"Camera data not found at {str(cam_pth)}"
intr_pth = data_dir / 'undistorted_intrinsics.json'
assert intr_pth.exists(), f"Intrinsic data not found at {str(intr_pth)}"
cam_data = load_cam_data(cam_pth, intr_pth)

extrinsics = {cam_id: torch.from_numpy(cam_data[cam_id]['c2w']) for cam_id in sorted(list(cam_data.keys()))}
intrinsics = {cam_id: torch.from_numpy(cam_data[cam_id]['intrinsics']) for cam_id in sorted(list(cam_data.keys()))}


# 0.2. get part proposal scores
part_score_fname = data_dir / "panoptic" / "part_scores.pkl"
assert part_score_fname.exists(), f"Part scores not found at {str(part_score_fname)}"
with open(part_score_fname, 'rb') as f:
    part_scores = pickle.load(f)

# 0.3. get node proposals
node_proposal_fname = data_dir / "panoptic" / "node_proposals.npy"
assert node_proposal_fname.exists(), f"Node proposals not found at {str(node_proposal_fname)}"
node_proposals = np.load(node_proposal_fname, allow_pickle=True).item()
node_proposals = {int(k): torch.tensor(v, dtype=torch.float32) for k, v in node_proposals.items()}

# 1.0 Do skeleton proposal
bones = get_bones()
skeletons, corresponding_2d_kpts = build_skeletons_with_dp_and_projection_filter(
    joint_pairs=bones,
    part_scores=part_scores,
    node_proposals=node_proposals,
    body_kpts_dict=body_kpts_dict,
    intrinsics=intrinsics,
    extrinsics=extrinsics,
    max_skeletons=max_skeleton,  # Limit to 2 skeletons for simplicity
)

print(f"Found {len(skeletons)} skeletons. ")
save_dir = data_dir / "panoptic"

# 1.1 Save skeleton bones
skel_meshes = dict()
for skel_idx, skel in enumerate(skeletons):
    for part_idx, (u, v) in enumerate(bones):
        if part_idx not in skel_meshes:
            skel_meshes[part_idx] = dict()

        if u in skel and v in skel:
            A = skel[u]
            B = skel[v]
            bone_mesh = create_bone_mesh(A, B)
            skel_meshes[part_idx][skel_idx] = dict(
                verts=bone_mesh.vertices,
                faces=bone_mesh.faces,
            )

np.save(save_dir / "viz_skel_proposals.npy", skel_meshes)

with open(save_dir / 'skeletons_proposals.pkl', 'wb') as f:
    pickle.dump(skeletons, f)

#################### DO NOT MODIFFY CODES ABOVE ####################
````

### 4.4. Joint Refinement

You should run this right after 4.3 as we do not have any feature saving ```corresponding_2d_kpts``` here.

````python

def refine_skeletons(skeletons, corresponding_2d_kpts, body_kpts_dict, intrinsics, extrinsics):
    """
    Refine 3D joints via 2D reprojection and least-squares fitting.

    input
      skeletons: List[Dict[joint_idx, np.array(3)]] list of clustered skeleton
      corresponding_2d_kpts: correspondence of skeleton 3D joints and 2d keypoint detection.
                define by yourself
      body_kpts_dict: Dict[cam_id, Tensor(N, J, 3)]
        - keypoint is [x, y, s] where s denotes score.
      extrinsics: Dict[cam_id, 4x4 tensor (cam to world)]
      intrinsics: Dict[cam_id, 3x3 tensor]
            - Note that this intrinsic is for image sized (1500, 2008). you should mathc them properly

    Returns:
        refined_skeletons: same format as input but with optimized joint positions
    """
    refined = []

    for skeleton, gt_kpt in zip(skeletons, corresponding_2d_kpts):
        refined_joints = dict()


        for joint_index, joint_3d in skeleton.items():

            if joint_index not in gt_kpt or len(gt_kpt[joint_index]) < 2:

                refined_joints[joint_index] = joint_3d
                continue

            A = []

            for cam_id, (det_index, kpt_2d) in gt_kpt[joint_index].items():

                K = intrinsics[cam_id].float().numpy()

                w2c = torch.linalg.inv(extrinsics[cam_id].float()).numpy()

                P = K @ w2c[:3, :]

                u_pixel, v_pixel = float(kpt_2d[0]), float(kpt_2d[1])

                A.append(u_pixel * P[2] - P[0])
                A.append(v_pixel * P[2] - P[1])

            A = np.array(A, dtype = np.float64)
            
            U, S, Vt = np.linalg.svd(A)

            X = Vt[-1]
            refined_joints[joint_index] = X[:3] / X[3]

        refined.append(refined_joints)

    return refined



#################### DO NOT MODIFFY CODES BELOW ####################

# 2.0. Do skeleton refinement
# with GT joint location, do refinement
refined_skeletons = refine_skeletons(
    skeletons=skeletons,
    corresponding_2d_kpts=corresponding_2d_kpts,
    body_kpts_dict=body_kpts_dict,
    intrinsics=intrinsics,
    extrinsics=extrinsics,
)

# 2.1. Save refined skeleton bones
skel_meshes = dict()
for skel_idx, skel in enumerate(refined_skeletons):
    for part_idx, (u, v) in enumerate(bones):
        if part_idx not in skel_meshes:
            skel_meshes[part_idx] = dict()

        if u in skel and v in skel:
            A = skel[u]
            B = skel[v]
            bone_mesh = create_bone_mesh(A, B)
            # Here you can save or visualize the bone mesh
            # For example, you can add it to a scene or save it to a file
            # scene.add_geometry(bone_mesh)
            # bone_mesh.export(f"bone_{u}_{v}_{part_idx}.stl")
            skel_meshes[part_idx][skel_idx] = dict(
                verts=bone_mesh.vertices,
                faces=bone_mesh.faces,
            )

np.save(save_dir / "refined_viz_skel_proposals.npy", skel_meshes)

with open(save_dir / 'refined_skeletons_proposals.pkl', 'wb') as f:
    pickle.dump(refined_skeletons, f)

print("Done!")
#################### DO NOT MODIFFY CODES ABOVE ####################
````
